```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd 
import numpy as np
from matplotlib.pyplot import plot
data_wasted_2003_2017 = pd.read_csv('datasets/wastestats.csv')
data_wasted_2003_2017.head()
```

```python
data_recy_2003_2017 = data_wasted_2003_2017[['year','waste_type','total_waste_recycled_tonne']].sort_values(by="year")
data_recy_2003_2017.reset_index()
data_recy_2003_2017.head()
```

```python
data_wasted_2018_2019 = pd.read_csv('datasets/2018_2019_waste.csv')
data_wasted_2018_2019.head()
```

```python
data_recy_2018_2019 = data_wasted_2018_2019 [['Year','Waste Type',"Total Recycled ('000 tonnes)"]].sort_values(by="Year")
data_recy_2018_2019.reset_index()
data_recy_2018_2019.columns = ["year","waste_type","total_waste_recycled_tonne"]
data_recy_2018_2019.total_waste_recycled_tonne = data_recy_2018_2019.total_waste_recycled_tonne * 1000.00
data_recy_2018_2019.head()
```

```python
data_energy = pd.read_csv('datasets/energy_saved.csv')
data_energy
```

```python
data_energy = pd.DataFrame({'waste_type':list(data_energy.iloc[2][1:]),
                           'energy_saved per 1 tonne':list(data_energy.iloc[3][1:]),
                           'crude_oil saved per 1 tonne':list(data_energy.iloc[4][1:])})
data_energy
```

```python
data_recy_2003_2019 =  pd.concat([data_recy_2003_2017,data_recy_2018_2019],ignore_index=True)
data_recy_2003_2019.head()
```

```python
data_recy_2003_2019.waste_type.unique()
```

```python
data_recy_2015_2019 = data_recy_2003_2019[(data_recy_2003_2019.year > 2014) & (data_recy_2003_2019.waste_type.isin(['Glass','Plastics','Plastic','Ferrous Metals','Ferrous Metal','Ferrous metal','Non-ferrous Metals','Non-ferrous metal','Non-ferrous metals','Non-Ferrous Metal']))]
data_recy_2015_2019
```

```python
data_recy_2015_2019.waste_type = data_recy_2015_2019.waste_type.replace('Plastics','Plastic')
data_recy_2015_2019.waste_type = data_recy_2015_2019.waste_type.replace(['Ferrous Metals','Ferrous metal'],'Ferrous Metal')
data_recy_2015_2019.waste_type = data_recy_2015_2019.waste_type.replace(['Non-ferrous Metals','Non-ferrous metals','Non-ferrous metal'],'Non-Ferrous Metal')
data_recy_2015_2019
```

```python
data_recy_energy = pd.merge(data_recy_2015_2019,data_energy,left_on ='waste_type',right_on='waste_type',how='left')
data_recy_energy = data_recy_energy[['year','waste_type','total_waste_recycled_tonne','energy_saved per 1 tonne']]
data_recy_energy['energy_saved per 1 tonne']=data_recy_energy['energy_saved per 1 tonne'].apply(lambda x:x.split(' ')[0])
data_recy_energy['energy_saved per 1 tonne']= pd.to_numeric(data_recy_energy['energy_saved per 1 tonne'])
data_recy_energy
```

```python
data_recy_energy['total_energy_saved'] = data_recy_energy.total_waste_recycled_tonne *  data_recy_energy['energy_saved per 1 tonne']
data_recy_energy
```

```python

annual_energy_savings = data_recy_energy.groupby('year').sum()[['total_energy_saved']]
annual_energy_savings

```

```python
annual_energy_savings.plot(kind="bar",xlabel="Year",title="Total energy by year")
```
